### 3. Analyzing the Content of a Categorical Variable

```python
obj_df = df.select_dtypes(include='object')
obj_cols = obj_df.columns

obj_cols
```

```python
#  function to see number of unique values in each categorical variables 

def describe_object(df, col_name):
    print(f"\nCOLUMN: {col_name}")
    print(f"{df[col_name].nunique()} different values")
```

```python
for col_name in obj_cols:
    describe_object(df, col_name)
    df[col_name].value_counts(dropna=False, normalize=True)
```

```python
#  view the top brands and product types

top_ten_brands=df['brand_name'].value_counts(dropna=False, normalize=True)[:10].index.tolist()
top_ten_product_type=df['product_type_1'].value_counts(dropna=False, normalize=True)[:10].index.tolist()[:9]
```

```python
top_ten_product_type
```

### 4. Summarizing Numerical Variables

```python
# now view the numerical variables

num_df = df.select_dtypes(include='number')
```

```python
num_cols = num_df.columns
num_cols
```

```python
# we will convert the datatypes of boolean values later
# view statistical information on numerical variables
# .describe() will work too

num_cols_no_boolean=['number_of_reviews', 'number_of_loves', 'price',
        'five_star', 'four_star',
       'three_star', 'two_star', 'one_star']
```

```python
def describe_numeric(df, col_name):
    print(f"\nCOLUMN: {col_name}")
    print(f"Minimum: {df[col_name].min()}")
    print(f"Maximum: {df[col_name].max()}")
    print(f"Average: {df[col_name].mean()}")
    print(f"Standard Deviation: {df[col_name].std()}")
    print(f"Median: {df[col_name].median()}")
```

```python
for col_name in num_cols:
    describe_numeric(df, col_name)
```

### 5. Visualizing the Data

We will be presenting each graph with Matplotlib and Seaborn

Visualize the distribution of a dataset:  
Histogram of Price, Number of Reviews, Number of Loves

```python
# Using matplotlib

fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, sharey=False)
plt.tight_layout()


ax1.hist(x=df.price)
ax1.set_xlabel('Price')
ax1.set_xticks(np.arange(0, 1000, 200))

ax2.hist(x=df.number_of_reviews)
ax2.set_xlabel('Number of Reviews')

ax3.hist(x=df.number_of_loves)
ax3.set_xlabel('Number of Loves')
```

```python
# Using seaborn

fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, sharey=False)
plt.tight_layout()


sns.distplot(df.price, ax=ax1)

sns.distplot(df.number_of_reviews, ax=ax2)

sns.distplot(df.number_of_loves, ax=ax3)
```

Visualize relationship between number of reviews and loves to Price:
Scatter plot of Number of Reviews, Number of Loves to Price

```python
# Using matplotlib
plt.subplot(1, 2, 1)
plt.scatter(x=df.number_of_reviews, y=df.number_of_loves)
plt.xlabel('Number of Loves')
plt.ylabel('Number of Reviews')

plt.subplot(1, 2, 2)
plt.scatter(x=df.price, y=df.number_of_loves)
plt.xlabel('Number of Loves')
plt.ylabel('Price')
plt.tight_layout()
```

```python
# Using Seaborn
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=False)
sns.scatterplot(x=df.number_of_reviews, y=df.number_of_loves,ax=ax1)
sns.scatterplot(x=df.price, y=df.number_of_loves, ax=ax2)
plt.tight_layout()
```

Visualizing how much average each product types or top 10 brands costs:  

Bar Chart of Product Types to Number of Loves

Bar Chart of Top Brands to Number of Loves

```python
# Using matplotlib
df_dropna=df.dropna()
```

```python
df_top_product_type=df.loc[df['product_type_1'].isin(top_ten_product_type)].dropna()
for product_type_1, df_product_type_1 in df_top_product_type.groupby('product_type_1'):
    avg_love=df_product_type_1.number_of_loves.mean()
    plt.bar(product_type_1, height=avg_love)
plt.xticks(rotation=90)
plt.ylabel('Number Of Loves')
plt.xlabel('Product Type')
```

```python
df_top_brands=df.loc[df['brand_name'].isin(top_ten_brands)]
for brand_1, df_brand_1 in df_top_brands.groupby('brand_name'):
    avg_love=df_brand_1.number_of_loves.mean()
    plt.bar(brand_1, height=avg_love)
plt.xticks(rotation=90)
plt.ylabel('Number Of Loves')
plt.xlabel('Top Brands')
```

```python
# Using Seaborn
df_top_product_type=df.loc[df['product_type_1'].isin(top_ten_product_type)].dropna()
chart=sns.barplot(x=df_top_product_type.product_type_1, y=df_top_product_type.number_of_loves)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
chart.set(xlabel='Product Types', ylabel='Number Of Loves')
```

```python
df_top_brands=df.loc[df['brand_name'].isin(top_ten_brands)]
chart=sns.barplot(x=df_top_brands.brand_name, y=df_top_brands.number_of_loves)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
chart.set(xlabel='Top 10 Brands', ylabel='Number Of Loves')
```
